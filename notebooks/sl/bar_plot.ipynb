### Bar plot

```python
"""
========
Barchart
========

A bar plot with errorbars and height labels on individual bars
"""
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

N = 5

# cropping
Cropping_means = (0.27846,0.280424,0.568071,0.620727,0.521819)
Cropping_std = (0.082569,0.08328,0.132937,0.100015,0.085396)

ind = np.arange(N)*2  # the x locations for the groups
width = 0.3       # the width of the bars

fig, ax = plt.subplots(figsize=(20,10))
rects1 = ax.bar(ind, Cropping_means, width, color='yellow', yerr=Cropping_std)
```

```python
# pasture

Pasture_means = (0.30024,0.302107,0.6249,0.646374,0.543684)

Pasture_std = (0.095359,0.095816,0.125614,0.095514,0.078893)

rects2 = ax.bar(ind + width, Pasture_means, width, color='mediumblue', yerr=Pasture_std)
```

```python
# Grass


Grass_means = (0.268001,0.27234,0.535515,0.640728,0.565116)

Grass_std = (0.075412,0.077327,0.147944,0.111985,0.099635)

rects3 = ax.bar(ind + width*2, Grass_means, width, color='lime', yerr=Grass_std)
```

```python
# Shurb


Shurb_means = (0.246353,0.251652,0.467262,0.631612,0.611322)

Shurb_std = (0.058537,0.05996,0.13538,0.102035,0.091714)

rects4 = ax.bar(ind + width*3, Shurb_means, width, color='orangered', yerr=Shurb_std)
```

```python
# Woodland

Woodland_means = (0.266581,0.270336,0.531126,0.620493,0.571735)

Woodland_std = (0.080781,0.082531,0.158684,0.131956,0.112166)

rects5 = ax.bar(ind + width*4, Woodland_means, width, color='violet', yerr=Woodland_std)
```

```python
# Forest

Forest_means = (0.279767,0.280927,0.527603,0.549035,0.506581)

Forest_std = (0.09448,0.094898,0.179039,0.167081,0.123869)


rects6 = ax.bar(ind + width*5, Forest_means, width, color='forestgreen', yerr=Forest_std)
```

```python

# add some text for labels, title and axes ticks
# ax.set_ylabel('Land cover type')
ax.set_ylabel('Average correlation coefficient',fontsize = 30)
ax.set_xticks(ind + width *3)
ax.set_xticklabels(('SPI_1M','SPI_3M','SPI_6M','SPI_12M','SPI_24M' ),fontsize = 30)

ax.legend((rects1[0], rects2[0],rects3[0], rects4[0],rects5[0], rects6[0]),
          ('Cropping','Pasture', 'Grass', 'Shurb', 'Woodland','Forest'),fontsize = 20)
```
