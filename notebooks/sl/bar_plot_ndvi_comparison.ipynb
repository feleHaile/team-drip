```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
NDVI_mean=[[0.27846,0.300246,0.268001,0.246353,0.266581,0.279767],
           [0.280424,0.302107,0.27234,0.251652,0.270336,0.280927],
           [0.568071,0.6249,0.535515,0.467262,0.531126,0.527603],
           [0.620727,0.646374,0.640728,0.631612,0.620493,0.549035],
           [0.521819,0.543684,0.565116,0.611322,0.571735,0.506581]]




```

```python

NDVI_STD=[[0.082569,0.095359,0.075412,0.058537,0.080781,0.09448],
         [0.08328,0.095816,0.077327,0.05996,0.082531,0.094898],
         [0.132937,0.125614,0.147944,0.13538,0.158684,0.179039],
         [0.100015,0.095514,0.111985,0.102035,0.131956,0.167081],
         [0.085396,0.078893,0.099635,0.091714,0.112166,0.123869]]






```

```python
columns = ('Crop','Pasture', 'Grass', 'Shurb', 'Woodland','Forest')
rows = ('SPI -1M','SPI -3M','SPI -6M','SPI -12M','SPI -24M' )

fig, ax = plt.subplots(figsize=(18,10))

# Get some pastel shades for the colors
colors = plt.cm.Set2(np.linspace(0, 0.5, len(rows)))
n_rows = len(NDVI_mean)

index = np.arange(len(columns))*2.5
bar_width = 0.4

# Initialize the vertical-offset for the stacked bar chart.
y_offset = np.zeros(len(columns))

# Plot bars and create text labels for the table
p=[]
for row in range(n_rows):
    p.append(plt.bar(index+bar_width*row, NDVI_mean[row], bar_width, bottom=y_offset, color=colors[row], align='center'))
    plt.errorbar(index+bar_width*row, NDVI_mean[row], yerr=NDVI_STD[row],fmt='.',color='black')
    
# Reverse colors and text labels to display the last value at the top.
```

```python
ax.set_ylabel('Average correlation coefficient',fontsize = 20)

ax.set_ylim(bottom=0, top= 0.9)
ax.legend((p[0][0], p[1][0],p[2][0],p[3][0],p[4][0]),rows,fontsize = 20)
ax.tick_params(labelsize=15)
ax.set_xticks(index + bar_width *2.5)
ax.set_xticklabels(columns,fontsize = 20)
ax.set_title('Comparison of Correlation (NDVI vs SPI) across Different Land Cover in MDB',fontsize=25)

plt.show()
```
